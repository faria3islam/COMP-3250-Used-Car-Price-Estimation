# Model Results Summary

This notebook provides a clean, presentation-ready view of model evaluation results.

It focuses on:
- core metrics (MAE, RMSE, R2)
- model comparison table
- final selected model

```python
import pandas as pd
from pathlib import Path

results_path = Path('../results/evaluation.csv')

if not results_path.exists() or results_path.stat().st_size == 0:
    raise FileNotFoundError(
        'No evaluation results found. Run: python ../src/main.py --skip-predict'
    )

metrics_df = pd.read_csv(results_path)
metrics_df
```

```python
display_cols = [c for c in ['model', 'mae', 'rmse', 'r2', 'n_train', 'n_test', 'n_samples'] if c in metrics_df.columns]
comparison = metrics_df[display_cols].sort_values(by=['rmse', 'mae'], ascending=True)
comparison.reset_index(drop=True)
```

```python
best_model = comparison.iloc[0]

print('Best model based on lowest RMSE:')
print(f"Model: {best_model.get('model', 'N/A')}")
print(f"MAE: {best_model.get('mae', float('nan')):.4f}")
print(f"RMSE: {best_model.get('rmse', float('nan')):.4f}")
print(f"R2: {best_model.get('r2', float('nan')):.4f}")
```
